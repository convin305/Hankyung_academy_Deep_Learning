# ResNet 응용

```python
from tensorflow.keras.applications.resnet50 import ResNet50, decode_predictions
```

```python
resnet = ResNet50()
resnet.summary()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = np.array(Image.open('images/dog.jpg').resize((224,224)))

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
yhat = resnet.predict(img.reshape(-1,224,224,3))            # 이미지는 항상 reshape를 해줘야 한다. 
label = decode_predictions(yhat)
label = label[0][0]
print(f'{label[1]} {label[2]*100:.2f}%')
```
