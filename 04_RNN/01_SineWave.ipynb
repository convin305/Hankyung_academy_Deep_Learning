# Simple RNN (Many to One 문제 )
### 사인파 시계열 예측 문제
- 사인파형 시계열을 입력으로 다음 스텝의 출력을 예측하기  
- 3개의 순서열을 임력한 후 출력값이 타겟과 일치하게 만든다. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s = np.sin(2*np.pi/8. * np.arange(20))
s
```

```python
plt.plot(s,'ro-')
plt.title("SineWave plot")
plt.grid()
plt.show()
```

### 원래의 시계열 벡터를 Toeplitz 행렬 형태로 변환.  
대각선 위의 성분들이 같은 정사각 행렬

```python
from scipy.linalg import toeplitz
S = np.fliplr(toeplitz(np.r_[s[-1],np.zeros(s.shape[0]-2)],s[::-1]))
S[:2]
```

```python
S.shape
```

```python
X_train = S[:-1,:3][:,:,np.newaxis]
y_train = S[:-1,3]
X_train.shape,y_train.shape
```

```python
plt.figure(figsize=(10,6))

plt.subplot(211)
plt.plot([0,1,2],X_train[0].flatten(),'bo-',label = 'input Sequence' )
plt.plot([3],y_train[0],'ro',label = 'target')
plt.xlim(-0.5,4.5)
plt.ylim(-1.1,1.1)
plt.legend()
plt.grid()
plt.title("First Sample Sequence")

plt.subplot(212)
plt.plot([1,2,3],X_train[1].flatten(),'bo-',label = 'input Sequence' )
plt.plot([4],y_train[1],'ro',label = 'target')
plt.xlim(-0.5,4.5)
plt.ylim(-1.1,1.1)
plt.legend()
plt.grid()
plt.title("Second Sample Sequence")

plt.tight_layout()
plt.show()
```

### 순환신경망 모델 정의 

```python
import tensorflow as tf
seed = 2021
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
```

```python
# input_shape=(3,1)
model = Sequential([
    SimpleRNN(10, input_shape=X_train.shape[1:], activation='tanh'),
    Dense(1,activation='linear')
])

# pip install numpy==1.19.5로 해야 돌아간다. 
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

### 모델 학습전

```python
plt.plot(y_train, 'ro-',label='target')
plt.plot(model.predict(X_train[:,:,:]),'bs-',label='predict')
plt.xlim(-0.5,20.5)
plt.ylim(-1.1,1.1)
plt.legend()
plt.title('before training')
plt.show()
```

### 모델 학습

```python
history = model.fit(X_train, y_train, epochs=100,verbose=0)
```

### 학습 결과 확인

```python
plt.plot(y_train, 'ro-',label='target')
plt.plot(model.predict(X_train[:,:,:]),'bs-',label='predict')
plt.xlim(-0.5,20.5)
plt.ylim(-1.1,1.1)
plt.legend()
plt.title('after training')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title("Loss")
plt.show()
```
